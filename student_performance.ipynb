```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "/content/ST IA-2 Marks A.xls"
df = pd.read_excel(file_path)
```

```python
marks = pd.to_numeric(df.iloc[3:, 5], errors='coerce')
marks = marks.dropna()
```

```python
topper = ((marks >= 40) & (marks <= 50)).sum()
medium = ((marks >= 20) & (marks <= 39)).sum()
average = ((marks >= 10) & (marks <= 19)).sum()
```

```python
print("Student Performance Analysis")
print("--------------------------------")
print("Topper Students (40-50):", topper)
print("Medium Students (20-39):", medium)
print("Average Students (10-19):", average)
```

```python
categories = ["Topper", "Medium", "Average"]
values = [topper, medium, average]
```

```python
plt.figure(figsize=(8,5))
plt.bar(categories, values)
```

```python
plt.title("Student Marks Analysis")
plt.xlabel("Student Category")
plt.ylabel("Number of Students")
```

```python
plt.show()
```
